<h1> Initialize Boring Stuffs </h1>

```python
#import libraries and stuff
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import time
from NeuralNetwork import Layer, NeuralNetwork
from PSO import ParticleStructure, Swarm

#import mnist from local directory
data = pd.read_csv('mnist/train.csv')

#data dimension measure
number_of_samples = data.shape[0]
number_of_features = data.shape[1] -1 

#data_target_split
Y = np.array(data['label'])
Y = np.reshape(Y, (number_of_samples, 1))
X = np.array(data[data.columns[1:]])
X = X / 255
X = X.T

#label vector regenerate
Y_Original = np.zeros((number_of_samples,10))
true_output_matrix = np.identity(10)
for i in range(number_of_samples):
    Y_Original[i] = Y_Original[i] + true_output_matrix[Y[i]]
Y = Y_Original.T
```

<h1> Neural Network with Backprop Run </h1> 

```python
#initialize network
PSO_Net = NeuralNetwork([2, 16, 10])
PSO_Net.structure_define()

num_epoch = 200
epoch = np.arange(1, num_epoch+1)
loss = np.zeros((num_epoch, 1))

tick = time.time()

for i in range(num_epoch):
    PSO_Net.forward_propagation(X)
    loss[i] = PSO_Net.calculate_network_loss(Y)
    print(loss[i])
    PSO_Net.backward_propagation(X, Y)
tock = time.time()

print("Runtime of Backprop: ", tock-tick)

plt.figure(1)
plt.semilogy(epoch, loss)
plt.xlabel('Epoch')
plt.ylabel('Error Amplitude in Log Dimension')
plt.show()
```

<h1> Particle Swarm Optimizer with NN </h1> 

```python
#initialize swarm
particleSwarm = Swarm(PSO_Net, X, Y, maxIt = 100, nPop = 50)
particleSwarm.initialize_swarm(PSO_Net)

#run main loop
particleSwarm.pso_loop(PSO_Net)
        
```
